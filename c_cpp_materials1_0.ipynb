````python
# Auto-set language metadata for all code cells in this notebook.
# Save notebook before running; it will rewrite the .ipynb file on disk.

import json
import re
from pathlib import Path

NB_PATH = Path("c_cpp_materials.ipynb")  # change if needed
MODE = "detect"  # "detect" -> set C/C++/Python; "auto" -> clear per-cell language metadata


def detect_language(src: str):
    lines = [l for l in src.splitlines() if l.strip()]
    if not lines:
        return None

    first = lines[0].lstrip()
    if first.startswith("%%"):
        tag = first[2:].split()[0].lower()
        if tag in ("cpp", "c++"):
            return "cpp"
        if tag == "c":
            return "c"
        if tag in ("python", "py"):
            return "python"

    text = "\n".join(lines)
    lower = text.lower()

    score = {"python": 0, "cpp": 0, "c": 0}

    py_tokens = [
        "import ", "from ", "def ", "print(", "lambda ",
        "none", "true", "false", "try:", "except", "with ",
        "yield ", "async ", "await ", "elif ", "self",
    ]
    cpp_tokens = [
        "#include", "std::", "using namespace std", "cout", "cin",
        "vector<", "string", "new ", "delete", "template<",
        "namespace ", "nullptr", "auto ", "::",
    ]
    c_tokens = [
        "#include", "printf(", "scanf(", "malloc(", "free(",
        "sizeof", "int main", "char *", "struct ", "typedef ",
        "enum ", "fopen(", "fgets(", "puts(", "gets(",
    ]

    for t in py_tokens:
        if t in lower or t in text:
            score["python"] += 2
    for t in cpp_tokens:
        if t in lower or t in text:
            score["cpp"] += 2
    for t in c_tokens:
        if t in lower or t in text:
            score["c"] += 2

    if "iostream" in lower or "std::" in lower or "using namespace std" in lower:
        score["cpp"] += 3
    if "stdio.h" in lower or "stdlib.h" in lower:
        score["c"] += 3

    if score["python"] > score["cpp"] and score["python"] > score["c"]:
        return "python"
    if score["cpp"] > score["c"]:
        return "cpp"
    if score["c"] > 0:
        return "c"

    if re.search(r"^\s*(def|class|for|while|if|elif|else|try|with)\b.*:\s*$", text, re.M):
        return "python"

    return None


nb = json.loads(NB_PATH.read_text(encoding="utf-8"))

updated = 0
cleared = 0

for cell in nb.get("cells", []):
    if cell.get("cell_type") != "code":
        continue
    src = "".join(cell.get("source", []))
    md = cell.setdefault("metadata", {})

    if MODE == "auto":
        if "language" in md:
            md.pop("language", None)
            cleared += 1
        vs = md.get("vscode")
        if isinstance(vs, dict) and "languageId" in vs:
            vs.pop("languageId", None)
            cleared += 1
            if not vs:
                md.pop("vscode", None)
        continue

    lang = detect_language(src)
    if lang is None:
        if "language" in md:
            md.pop("language", None)
            cleared += 1
        vs = md.get("vscode")
        if isinstance(vs, dict) and "languageId" in vs:
            vs.pop("languageId", None)
            cleared += 1
            if not vs:
                md.pop("vscode", None)
        continue

    display, lang_id = {
        "cpp": ("C++", "cpp"),
        "c": ("C", "c"),
        "python": ("python", "python"),
    }[lang]

    if md.get("language") != display:
        md["language"] = display
        updated += 1

    vs = md.get("vscode")
    if not isinstance(vs, dict):
        vs = {}
    if vs.get("languageId") != lang_id:
        vs["languageId"] = lang_id
        md["vscode"] = vs
        updated += 1

NB_PATH.write_text(json.dumps(nb, ensure_ascii=False, indent=1), encoding="utf-8")
print(f"Updated: {updated}, Cleared: {cleared}")
````

## Chương 1. Cơ bản: cấu trúc chương trình và xuất dữ liệu

**Prerequisites**
- Không yêu cầu.

**Learning outcomes**
- Nhận biết được “khung” tối thiểu của một chương trình C/C++ khi làm bài thi (`#include`, `main`, `{}`, `;`).
- Biết dùng comment để chú thích và dùng `printf` để in ra màn hình.
- Hiểu các escape sequence cơ bản trong chuỗi: `\n`, `\t`, `\"`.

### Giải thích

#### 1) “Khung” tối thiểu của chương trình (khi viết file .c/.cpp)
Khi làm bài thi hoặc viết chương trình dưới dạng file, bạn thường gặp dạng chuẩn sau:

```c
#include <stdio.h>

int main(void) {
    printf("Hello!\n");
    return 0;
}
```

Giải thích nhanh từng phần:
- `#include <stdio.h>`: nạp thư viện I/O chuẩn để dùng `printf`.
- `int main(void)`: hàm **bắt đầu chương trình**. `void` nghĩa là không nhận tham số.
- `{ ... }`: **khối lệnh**. Các câu lệnh nằm trong khối sẽ được thực thi theo thứ tự.
- `;`: kết thúc **một câu lệnh** trong C.

> Ghi chú cho notebook (xcpp17/xeus-cling): trong notebook, bạn có thể chạy trực tiếp các câu lệnh như `printf(...)` mà **không cần** `main`. Để vừa “đúng form đề thi” vừa chạy được, các ví dụ dưới đây sẽ **comment** dòng `int main...{` và `}` (và thường comment cả `return 0;`).

#### 2) In ra màn hình với `printf`
- `printf` dùng để in ra màn hình.
- Ở Chương 1, bạn chỉ cần in **chuỗi** và dùng `\n` để xuống dòng.

Bẫy hay gặp (đề thi rất hay hỏi):
- Quên dấu `;` ở cuối câu lệnh → lỗi biên dịch.
- Quên `#include <stdio.h>` → không dùng được `printf` (hoặc bị warning).
- Nhầm `'A'` (ký tự) với `"A"` (chuỗi).

#### 3) Escape sequence trong chuỗi
Escape sequence là “ký tự đặc biệt” bắt đầu bằng `\` trong chuỗi:
- `\n`: xuống dòng.
- `\t`: tab.
- `\"`: in dấu nháy kép `"` bên trong chuỗi.

#### 4) Comment
- `// comment một dòng`.
- `/* comment nhiều dòng */`.

Bẫy: quên đóng `*/` sẽ khiến phần sau bị “nuốt” thành comment.

### Ví dụ

#### Ví dụ 1 — Hello + khung chương trình
- Mục tiêu: thấy “form đề thi” có `main`, nhưng vẫn chạy được trong notebook.

- Bẫy cần nhớ: `\n` để xuống dòng, và dấu `;` ở cuối câu lệnh.


````python
#include <stdio.h>

// Trong file .c/.cpp khi làm bài thi, chương trình thường bắt đầu ở main:
//int main(void) {
{
    // Mọi câu lệnh trong C kết thúc bằng dấu ';'
    printf("Hello C!\n");

    // In thêm 1 dòng để thấy tác dụng của '\n' (xuống dòng)
    printf("Dong 2: Chay trong notebook (khong can main).\n");

    // Trong file .c/.cpp: return 0; báo kết thúc thành công.
    // Trong notebook: 'return' không đứng ngoài hàm, nên ta comment.
    //return 0;
}
//} // Kết thúc main
````

#### Ví dụ 2 — Escape sequence: \n, \t, \"

- Kết quả kỳ vọng: các cột thẳng hàng bằng tab, và in được dấu nháy kép.

## Chương 2. Biến, kiểu dữ liệu, hằng, sizeof, định dạng in

**Prerequisites**
- Chương 1.

**Learning outcomes**
- Khai báo và khởi tạo biến cơ bản (`int`, `double`, `char`).
- Hiểu chia nguyên vs chia thực và biết ép kiểu để ra kết quả mong muốn.
- Biết dùng `const` để tạo hằng và dùng `sizeof` để biết kích thước theo byte.
- In đúng định dạng với `printf` (bẫy rất hay gặp trong đề).

### Giải thích

#### 1) Biến là gì? (và vì sao phải khởi tạo)
- **Biến** là “một vùng nhớ có tên” dùng để lưu giá trị.
- Bạn nên **khởi tạo** biến ngay khi khai báo để tránh “giá trị rác”.

Ví dụ:
- `int age = 20;` (đã khởi tạo)
- `int x;` rồi dùng `x` ngay (đề thi hay gài) → giá trị **không xác định**.

#### 2) Kiểu dữ liệu cơ bản và ý nghĩa
- `int`: số nguyên.
- `double` (hoặc `float`): số thực.
- `char`: ký tự, nhưng thực chất cũng là **một số nguyên nhỏ** (mã ASCII).

Bẫy hay gặp:
- Chia nguyên: `5/2` cho ra `2` (không phải `2.5`).
- In sai định dạng `printf` (ví dụ dùng `%d` cho `double`) → kết quả sai/khó đoán.

#### 3) Hằng với `const`
- `const` tạo ra biến “không đổi” sau khi khởi tạo.
- Nếu cố gắng gán lại sẽ bị lỗi biên dịch.

#### 4) `sizeof` trả về gì?
- `sizeof(T)` cho biết **số byte** của kiểu `T`.
- Giá trị trả về có kiểu `size_t` (thường là số nguyên không âm).
- Khi in `sizeof`, dùng `%zu` (đúng chuẩn).

#### 5) `printf` format cơ bản (mức thi)
Một số format hay dùng:
- `%d`: `int`
- `%f`: số thực (`double` khi truyền vào `printf`)
- `%.2f`: in số thực với 2 chữ số sau dấu phẩy
- `%c`: `char`
- `%zu`: `size_t` (kết quả của `sizeof`)

### Ví dụ

#### Ví dụ 1 — Khai báo biến + in đúng định dạng
- Mục tiêu: nhìn format `printf` và hiểu “in cái gì thì dùng % gì”.
- Bẫy cần nhớ: `%d` ≠ `%f` ≠ `%c`.


````python
#include <stdio.h>

// Nếu viết file .c/.cpp, các dòng dưới thường nằm trong main:
//int main(void) {
{
    int age = 20;
    double pi = 3.14159;
    char grade = 'A';
    const int YEAR = 2026;

    printf("Age = %d\n", age);
    printf("Pi = %.2f\n", pi);
    printf("Grade = %c\n", grade);
    printf("YEAR = %d\n", YEAR);

    // Sai format (đã comment): dùng %d cho double có thể in ra giá trị rác.
    //printf("Pi (sai format) = %d\n", pi);
}
//}
````

#### Ví dụ 2 — Chia nguyên vs chia thực (bẫy kinh điển)
- Mục tiêu: dự đoán đúng kết quả của `5/2` và biết cách ép kiểu để ra `2.5`.
- Kết quả kỳ vọng: dòng đầu ra `2`, các dòng sau ra `2.500000`.


````python
#include <stdio.h>

{
    int a = 5;
    int b = 2;

    // a và b đều là int => chia nguyên
    printf("a/b (int) = %d\n", a / b);

    // Ép kiểu để phép chia trở thành chia thực
    printf("(double)a/b = %f\n", (double)a / b);
    printf("a/(double)b = %f\n", a / (double)b);
}
````

#### Ví dụ 3 — `char` là ký tự nhưng cũng là số (ASCII)
- Mục tiêu: thấy rõ sự khác nhau giữa `%c` và `%d` khi in một `char`.
- Kết quả kỳ vọng: in `A` và in mã ASCII (65).


````python
#include <stdio.h>

{
    char ch = 'A';

    printf("ch as char = %c\n", ch);
    printf("ch as ASCII = %d\n", (int)ch);
}
````

#### Ví dụ 4 — `sizeof` và cách in an toàn với `%zu`
- Mục tiêu: biết `sizeof` trả về số byte, và dùng `%zu` để in đúng.
- Ghi nhớ: `sizeof` không phải hàm; nó là toán tử.


````python
#include <stdio.h>

{
    printf("sizeof(char)   = %zu byte\n", sizeof(char));
    printf("sizeof(int)    = %zu bytes\n", sizeof(int));
    printf("sizeof(double) = %zu bytes\n", sizeof(double));
}
````

### Luyện tập

2. Khai báo một ký tự và in ra mã ASCII của nó (dùng `%d`).
3. Viết hằng `PI = 3.14` và dùng để tính chu vi hình tròn bán kính 5.

<details>
<summary>Solution</summary>

Gợi ý:
- Diện tích hình chữ nhật: `area = length * width`.
- Ép kiểu `int` khi in mã ASCII: `printf("%d", (int)ch);`.
- Chu vi: `2 * PI * r`.
</details>


### Lỗi thường gặp
- Dùng sai định dạng `printf` (ví dụ `%d` cho `double`).
- Quên khởi tạo biến trước khi dùng.

## Chương 3. Toán tử và biểu thức

**Prerequisites**
- Chương 2.

**Learning outcomes**
- Dự đoán đúng kết quả biểu thức số học (đặc biệt: **chia nguyên** và `%`).
- Phân biệt rõ `=` (gán) và `==` (so sánh) trong điều kiện.
- Hiểu `&&`, `||`, `!` và cơ chế **short-circuit** (dừng sớm) để tránh lỗi.
- Dùng đúng toán tử `?:` và phân biệt `++i` với `i++` trong biểu thức.
- Biết các bẫy phổ biến trong đề: chia nguyên, nhầm toán tử, viết biểu thức gây “không xác định”.

### Giải thích

#### 1) Toán tử số học: `+ - * / %`
- `+ - * /`: phép toán cơ bản.
- `/` có 2 “kiểu chia” (đề thi rất hay hỏi):
  - Nếu **hai toán hạng đều là `int`** → **chia nguyên** (bỏ phần thập phân).
  - Nếu **có ít nhất 1 toán hạng là số thực** (`float/double`) → **chia thực**.
- `%` (modulo/phần dư) chỉ dùng cho **số nguyên**.

Bẫy kinh điển:
- `7/2` cho ra `3` (không phải `3.5`).
- `7%2` cho ra `1`.

#### 2) `=` (gán) khác `==` (so sánh)
- `=`: **thay đổi biến** và biểu thức gán cũng có **giá trị** (giá trị vừa gán).
- `==`: **so sánh** và cho ra đúng/sai.

Bẫy rất hay gặp:
- `if (x = 0)` là **gán**, và điều kiện sẽ thành **false** (vì gán 0).
- `if (x = 5)` là **gán**, và điều kiện sẽ thành **true** (vì gán 5 ≠ 0).

#### 3) Toán tử logic `&& || !` và short-circuit
- `A && B`: nếu `A` **false** thì **không cần** kiểm tra `B` nữa. (And - Kết quả `true` chỉ khi cả `A` và `B` đều `true`.)
- `A || B`: nếu `A` **true** thì **không cần** kiểm tra `B` nữa. (Or - Chỉ `false` khi cả `A` và `B` đều `false`.)

- `!A`: đảo giá trị logic của `A` (NOT) — nếu `A` true thì `!A` false, nếu `A` false thì `!A` true

Ứng dụng rất thực tế (dạng đề hay ra):
- Tránh chia cho 0: `a != 0 && (b/a > 1)`.

#### 4) Toán tử điều kiện `?:`

- `condition ? expr_if_true : expr_if_false`

Dạng: `cond ? a : b`
- Nếu `cond` đúng → lấy `a`.
- Nếu `cond` sai → lấy `b`.

Hay dùng để gán nhanh:
- `max = (a > b) ? a : b;`

Bẫy: biểu thức dài nên thêm ngoặc để dễ đọc và tránh nhầm ưu tiên.

#### 5) `++i` vs `i++` và bẫy “không xác định”
- `++i` (tiền tố): **tăng trước**, rồi mới dùng.
- `i++` (hậu tố): **dùng trước**, rồi mới tăng.

Bẫy mức thi (đừng viết):
- `i = i++;` hoặc `i++ + ++i` → kết quả **không đáng tin** (không nên suy luận).

#### 6) Thứ tự ưu tiên toán tử (cần nhớ mức tối thiểu)
- `* / %` ưu tiên hơn `+ -`.
- So sánh (`==`, `!=`, `<`, ...) thường được nhóm trước logic `&&`, `||`.
- Gán (`=`) thường gần cuối.

> Mẹo làm bài thi: nếu thấy “lăn tăn” khi đọc biểu thức, **tự thêm ngoặc** khi phân tích.

## Chương 4. Điều khiển luồng

**Prerequisites**

- Phân biệt `for/while/do-while` (đặc biệt: `do-while` chạy ít nhất 1 lần).
- Dự đoán output khi có `break/continue`.

### Giải thích

#### 1) Điều kiện trong C/C++: 0 và khác 0
Trong C/C++:
- `0` được hiểu là **false**.
- Bất kỳ giá trị khác `0` được hiểu là **true**.

Vì vậy `if (x)` nghĩa là “nếu x khác 0”.

Bẫy: `if (x = 5)` là **gán**, và vì 5 ≠ 0 nên luôn true.

#### 2) `if / else if / else` và bẫy “else gắn với if gần nhất”
Nếu không dùng `{}` thì `else` sẽ gắn với `if` gần nhất (đề thi hay gài).

Mẹo an toàn khi làm bài:
- Nếu trong `if` có nhiều hơn 1 câu lệnh, hãy luôn dùng `{}`.

#### 3) `switch/case`: fall-through và `break`
- `switch` phù hợp khi so sánh một biến với **nhiều giá trị rời rạc**.
- Nếu thiếu `break`, chương trình sẽ **rơi qua** (fall-through) sang `case` tiếp theo.

#### 4) Vòng lặp `for / while / do-while`
- `for`: tiện khi biết trước số lần lặp.
- `while`: lặp khi còn thỏa điều kiện.
- `do-while`: **chạy trước, kiểm tra sau** → luôn chạy ít nhất 1 lần.

Bẫy rất hay gặp: off-by-one (`i < n` khác `i <= n`).

#### 5) `break` và `continue`
- `break`: thoát khỏi **vòng lặp hiện tại** (hoặc thoát `switch`).
- `continue`: bỏ qua phần còn lại của lần lặp hiện tại và sang lần lặp tiếp theo.

> Lưu ý: trong vòng lặp lồng nhau, `break` chỉ thoát **1 mức**.

### Ví dụ

#### Ví dụ 1 — `switch` thiếu `break` (fall-through)
- Mục tiêu: nhìn output để “khắc” vào đầu: thiếu `break` sẽ chạy tiếp case sau.


````python
#include <stdio.h>

{
    int day = 2;

    printf("Switch thieu break:\n");
    switch (day) {
        case 2:
            printf("Thu 2\n");
            // THIEU break => rơi qua case 3
        case 3:
            printf("Thu 3\n");
            break;
        default:
            printf("Khac\n");
            break;
    }

    printf("\nSwitch day du break:\n");
    switch (day) {
        case 2:
            printf("Thu 2\n");
            break;
        case 3:
            printf("Thu 3\n");
            break;
        default:
            printf("Khac\n");
            break;
    }
}
````

#### Ví dụ 2 — `while` vs `do-while`
- Mục tiêu: thấy `do-while` chạy ít nhất 1 lần, dù điều kiện ban đầu sai.


````python
#include <stdio.h>

{
    int i = 0;

    printf("while (i < 0):\n");
    while (i < 0) {
        printf("i=%d\n", i);
        i++;
    }
    printf("Ket thuc while (khong in gi vi dieu kien sai ngay tu dau)\n\n");

    i = 0;
    printf("do {..} while (i < 0):\n");
    do {
        printf("i=%d\n", i);
        i++;
    } while (i < 0);
    printf("Ket thuc do-while (van chay 1 lan)\n");
}
````

#### Ví dụ 3 — Off-by-one trong `for` (`i < n` vs `i <= n`)
- Mục tiêu: nhìn số lần lặp để tránh lỗi vượt biên mảng.
- Gợi ý: với mảng `a[n]`, chỉ số hợp lệ là `0..n-1`.


````python
#include <stdio.h>

{
    int n = 5;

    printf("for (i=0; i<n; i++): ");
    for (int i = 0; i < n; i++) {
        printf("%d ", i);
    }
    printf("\n");

    printf("for (i=0; i<=n; i++): ");
    for (int i = 0; i <= n; i++) {
        printf("%d ", i);
    }
    printf("\n");

    // Nếu i dùng làm chỉ số mảng a[n], vòng i<=n sẽ cố truy cập a[n] (SAI).
}
````

#### Ví dụ 4 — `break` và `continue` trong vòng lặp
- Mục tiêu: dự đoán output: `continue` bỏ qua, `break` thoát sớm.


````python
#include <stdio.h>

{
    for (int i = 1; i <= 10; i++;) {
        if (i == 5) {
            // Bỏ qua số 5
            continue;
        }
        if (i == 8) {
            // Dừng vòng lặp khi gặp 8
            break;
        }
        printf("%d ", i);
    }
    printf("\n");
}
````

### Giải thích

- **Prototype** (khai báo): cho compiler biết “hàm tên gì, nhận gì, trả gì” để có thể gọi.

#### 3) Scope & lifetime
- Biến cục bộ: chỉ sống trong block `{ ... }`.
- Biến global: sống suốt chương trình.
- `static` cục bộ: scope như biến cục bộ nhưng **giữ giá trị** qua nhiều lần gọi hàm.

Bẫy: trả về địa chỉ của biến cục bộ (sẽ gặp lại khi học con trỏ/malloc).

#### 4) Recursion (đệ quy) mức cơ bản
- Hàm gọi lại chính nó.

#### Ví dụ 2 — Swap “không đổi” (pass-by-value) vs swap đúng bằng pointer
- Mục tiêu: bẫy thi kinh điển: đổi trong hàm nhưng ngoài không đổi.
- Gợi mở: pointer là “biến lưu địa chỉ”. `&x` là địa chỉ của x, `*p` là giá trị tại địa chỉ p.


````python
#include <stdio.h>

void swap_by_value(int a, int b);
void swap_by_ptr(int* a, int* b);

{
    int x = 1;
    int y = 2;

    printf("Ban dau: x=%d y=%d\n", x, y);

    swap_by_value(x, y);
    printf("Sau swap_by_value: x=%d y=%d (KHONG DOI)\n", x, y);

    swap_by_ptr(&x, &y);
    printf("Sau swap_by_ptr:   x=%d y=%d (DA DOI)\n", x, y);
}

#ifndef CH5_EX2_SWAP_DEFINED
#define CH5_EX2_SWAP_DEFINED
void swap_by_value(int a, int b) {
    int tmp = a;
    a = b;
    b = tmp;
    // a,b chỉ là bản sao => không ảnh hưởng biến gốc
}

void swap_by_ptr(int* a, int* b) {
    // a,b là con trỏ: a trỏ tới biến x, b trỏ tới biến y
    int tmp = *a; // lấy GIÁ TRỊ tại địa chỉ a
    *a = *b;
    *b = tmp;
}
#endif
````

## Chương 7. Địa chỉ bộ nhớ và con trỏ cơ bản

**Prerequisites**
- Chương 2 (kiểu dữ liệu).
- Chương 5 (hàm — để hiểu ví dụ swap).

**Learning outcomes**
- Hiểu “địa chỉ bộ nhớ” là gì và vì sao biến nào cũng có địa chỉ.
- Nắm chắc 3 ký hiệu quan trọng: `&` (lấy địa chỉ), `*` (giải tham chiếu), `NULL`.
- Viết được đoạn code dùng con trỏ để **đọc/ghi** giá trị của biến thông qua địa chỉ.
- Hiểu vì sao `scanf/sscanf` cần truyền **địa chỉ**.

### Giải thích

#### 1) Địa chỉ bộ nhớ là gì?
Hãy tưởng tượng mỗi biến là một “hộp” trong RAM. Hộp này có:
- **Giá trị** (value) đang chứa.
- **Địa chỉ** (address) — giống như “số nhà” của hộp.

#### 2) Con trỏ (pointer) là gì?
- Con trỏ là biến dùng để **lưu địa chỉ** của biến khác.
- Ví dụ: `int *p = &x;` nghĩa là `p` đang giữ địa chỉ của `x`.

Ký hiệu quan trọng:
- `&x`: lấy địa chỉ của `x`.
- `*p`: lấy **giá trị tại địa chỉ** mà `p` đang trỏ tới (gọi là giải tham chiếu).

#### 3) `NULL` và an toàn khi dereference
- `NULL` (hoặc `nullptr` trong C++) biểu thị “không trỏ tới đâu”.
- Dereference `*p` khi `p == NULL` → lỗi runtime (crash/UB).

Thói quen tốt:
- Khởi tạo con trỏ về `NULL` nếu chưa trỏ tới đâu.
- Luôn check `p != NULL` trước khi dùng `*p`.

#### 4) Vì sao `scanf` cần địa chỉ?
`scanf/sscanf` phải ghi kết quả vào biến của bạn, nên nó cần biết **nơi để ghi** → chính là **địa chỉ**.

Bẫy thi cực hay gặp:
- `scanf("%d", a);` là sai (vì `a` là giá trị, không phải địa chỉ).
- Đúng: `scanf("%d", &a);`

### Ví dụ

#### Ví dụ 1 — Trực quan hóa `&` và `*`
- Mục tiêu: nhìn rõ 3 thứ: `&x` (địa chỉ), `p` (địa chỉ), và `*p` (giá trị).
- Lưu ý: địa chỉ in ra có thể khác nhau mỗi lần chạy.


````python
#include <stdio.h>

{
    int x = 10;
    int* p = &x; // p lưu địa chỉ của x

    printf("Gia tri x = %d\n", x);

    // %p dùng để in địa chỉ; cast về (void*) để đúng format
    printf("Dia chi &x = %p\n", (void*)&x);
    printf("Gia tri p  = %p\n", (void*)p);

    printf("Gia tri *p = %d\n", *p);

    // Ghi thông qua con trỏ: đổi giá trị của x
    *p = 20;
    printf("Sau *p = 20 -> x = %d\n", x);
}
````

## Chương 8. Con trỏ & mảng, con trỏ hàm

**Prerequisites**
- Chương 6 (mảng).
- Chương 7 (con trỏ).

**Learning outcomes**
- Hiểu quy tắc “array decay”: trong nhiều biểu thức, tên mảng `a` được dùng như địa chỉ `&a[0]`.
- Duyệt mảng bằng con trỏ và phân biệt `*(p+i)` với `*p+i`.
- Phân biệt `a`, `&a`, `&a[0]` (bẫy hay gặp).
- Khai báo và dùng con trỏ hàm: `int (*pf)(int,int)`.
- Dùng mảng con trỏ hàm để tạo “menu” chọn hành vi.

### Giải thích

#### 1) Mảng và con trỏ liên quan thế nào?
Với mảng `int a[4]`:
- `a[i]` tương đương `*(a + i)`.
- `a` thường được hiểu như “địa chỉ phần tử đầu” (`&a[0]`).

Bẫy rất hay gặp:
- `*(p+i)` khác `*p+i` do ưu tiên toán tử.

#### 2) `a` vs `&a` vs `&a[0]`
- `a` và `&a[0]` có cùng **giá trị địa chỉ**, nhưng kiểu khác nhau trong một số ngữ cảnh.
- `&a` là “con trỏ tới cả mảng, là địa chỉ ngay sau mảng” (tăng `&a + 1` sẽ nhảy theo **kích thước cả mảng**).

#### 3) Con trỏ hàm
- Con trỏ hàm lưu “địa chỉ của hàm”.
- Cú pháp quan trọng: dấu ngoặc `(*pf)`.
  - Đúng: `int (*pf)(int,int);`
  - Sai (nhầm): `int *pf(int,int);` (đây là **hàm trả về con trỏ**).

### Ví dụ

#### Ví dụ 1 — Duyệt mảng bằng con trỏ (pointer arithmetic)

## Chương 10. Cấp phát bộ nhớ động

**Prerequisites**
- Chương 7–8 (con trỏ).

**Learning outcomes**
- Biết dùng `malloc/calloc/realloc/free` theo pattern chuẩn (check NULL, free đúng 1 lần).
- Hiểu stack vs heap và vì sao heap phải tự `free`.
- Nhận diện lỗi thường gặp: memory leak, double free, use-after-free.
- Biết cấp phát chuỗi động đúng quy tắc `n + 1` cho `\0`.

### Giải thích

#### 1) Stack vs heap
- **Stack**: biến cục bộ (trong hàm) được cấp phát tự động và thu hồi khi ra khỏi scope.
- **Heap**: vùng nhớ động — bạn tự cấp phát và tự giải phóng.

#### 2) `malloc/calloc/realloc/free`
- `malloc(nBytes)`: cấp phát `nBytes` byte (không khởi tạo).
- `calloc(count, size)`: cấp phát và khởi tạo 0.
- `realloc(p, newSize)`: đổi kích thước vùng nhớ.
- `free(p)`: giải phóng.

Pattern chuẩn (đề rất hay hỏi):
- Luôn check `p != NULL` sau `malloc/calloc/realloc`.
- Khi `realloc`, nên dùng biến tạm:
  - `tmp = realloc(p, newSize); if(tmp) p = tmp; else giữ p cũ`.

> Ghi chú: notebook đang chạy C++17. Trong C++ cần cast kết quả `malloc` về đúng kiểu. Trong C thuần thì không cần cast.

#### 3) Lỗi thường gặp
- Memory leak: quên `free`.
- Double free: `free` cùng con trỏ 2 lần (UB).
- Use-after-free: `free` xong vẫn truy cập.

Thói quen tốt: `free(p); p = NULL;`

#### 4) Cấp phát chuỗi động
Chuỗi C cần `\0` ở cuối, nên nếu muốn chứa `n` ký tự thì phải cấp phát `n+1`.

### Ví dụ

#### Ví dụ 1 — `malloc` mảng int + tính tổng + `free`
- Mục tiêu: pattern chuẩn: `n * sizeof(int)` + check NULL + free.


````python
#include <stdio.h>
#include <stdlib.h>

{
    int n = 5;

    // Trong C++ cần cast; trong C thuần thì không cần.
    int* a = (int*)malloc(n * sizeof(int));
    if (a == NULL) {
        printf("malloc failed\n");
    } else {
        for (int i = 0; i < n; i++) {
            a[i] = i + 1;
        }

        int sum = 0;
        for (int i = 0; i < n; i++) {
            sum += a[i];
        }
        printf("sum = %d\n", sum);

        free(a);
        a = NULL;
    }
}
````

#### Ví dụ 2 — `realloc` an toàn với biến tạm
- Mục tiêu: tránh mất con trỏ khi `realloc` thất bại.


````python
#include <stdio.h>
#include <stdlib.h>

{
    int n = 3;
    int* p = (int*)malloc(n * sizeof(int));
    if (!p) {
        printf("malloc failed\n");
    } else {
        for (int i = 0; i < n; i++) p[i] = i + 1;

        int newN = 6;
        int* tmp = (int*)realloc(p, newN * sizeof(int));
        if (!tmp) {
            // p vẫn còn hợp lệ, ta phải free p trước khi kết thúc
            printf("realloc failed, p con hop le\n");
            free(p);
        } else {
            p = tmp;

            for (int i = n; i < newN; i++) p[i] = i + 1;

            int sum = 0;
            for (int i = 0; i < newN; i++) sum += p[i];
            printf("sum(new) = %d\n", sum);

            free(p);
        }
    }
}
````

#### Ví dụ 3 — `free` + set `NULL` để tránh double free
- Mục tiêu: nhớ thói quen giúp tránh lỗi nghiêm trọng.


````python
#include <stdio.h>
#include <stdlib.h>

{
    int* p = (int*)malloc(sizeof(int));
    if (!p) {
        printf("malloc failed\n");
    } else {
        *p = 42;
        printf("*p = %d\n", *p);

        free(p);
        p = NULL;

        // free(NULL) là an toàn.
        free(p);
        printf("freed safely\n");

        // DANG SAI (comment): dùng sau free
        //printf("%d\n", *p);
    }
}
````

#### Ví dụ 4 — Chuỗi động: cấp phát `n+1` cho `\0`
- Mục tiêu: tránh quên null-terminator.

### Ví dụ

#### Ví dụ 1 — `typedef struct` + khởi tạo + in dữ liệu

````python
#include <stdio.h>

#ifndef CH11_TYPES_DEFINED
#define CH11_TYPES_DEFINED

typedef struct {
    char name[30];
    double gpa;
} Student;

#endif

{
    // Khởi tạo nhanh (cú pháp C/C++ đều chấp nhận)
    Student s = {"An", 3.5};

    printf("Name: %s\n", s.name);
    printf("GPA : %.2f\n", s.gpa);
}
````

#### Ví dụ 2 — Con trỏ struct: `->` và `(*p).x`
- Mục tiêu: đọc đề dạng con trỏ struct.


````python
#include <stdio.h>

#ifndef CH11_TYPES_DEFINED
#define CH11_TYPES_DEFINED

typedef struct {
    char name[30];
    double gpa;
} Student;

#endif

{
    Student s = {"Binh", 2.8};
    Student* p = &s;

    // Hai cách viết tương đương:
    p->gpa = 3.2;
    (*p).gpa = 3.4;

    printf("%s GPA=%.1f\n", p->name, p->gpa);
}
````

#### Ví dụ 3 — `enum` + `switch`
- Mục tiêu: hiểu enum là số nguyên có tên; dùng switch rõ nghĩa.


````python
#include <stdio.h>

#ifndef CH11_ENUM_DEFINED
#define CH11_ENUM_DEFINED

enum Level { LOW = 1, MEDIUM, HIGH };

#endif

{
    enum Level lv = MEDIUM;

    switch (lv) {
        case LOW:
            printf("LOW\n");
            break;
        case MEDIUM:
            printf("MEDIUM\n");
            break;
        case HIGH:
            printf("HIGH\n");
            break;
        default:
            printf("UNKNOWN\n");
            break;
    }

    // In enum như int (vì bản chất enum là số nguyên)
    printf("MEDIUM = %d\n", MEDIUM);
}
````

### Giải thích

#### 1) `FILE` và `FILE*` là gì?
- `FILE` là kiểu dữ liệu do thư viện chuẩn cung cấp, đại diện cho **trạng thái của một luồng file (file stream)**.
- `FILE* fp` là **con trỏ** trỏ tới đối tượng stream đó.

Vì sao dùng con trỏ?
- `fopen` cần trả về “một stream đã được mở”, và nếu mở thất bại thì trả về `NULL`.

#### 2) Quy trình chuẩn khi làm bài
1. `FILE* fp = fopen(...);`
2. `if (!fp) { ... }`
3. Đọc/ghi (`fprintf`, `fgets`, ...)

### Ví dụ

#### Ví dụ 1 — Ghi file text (mode `w`) + check NULL
- Mục tiêu: pattern chuẩn: fopen → check → fprintf → fclose.


````python
#include <stdio.h>

{
    const char* path = "ch12_ex1.txt";
    FILE* fp = fopen(path, "w");
    if (!fp) {
        printf("Khong mo duoc file\n");
    } else {
        fprintf(fp, "Dong 1\n");
        fprintf(fp, "Dong 2\n");

        fclose(fp);
        printf("Da ghi xong: %s\n", path);
    }
}
````

#### Ví dụ 2 — `w` (ghi đè) vs `a` (append)
- Mục tiêu: nhớ mode nào ghi thêm, mode nào ghi đè.


````python
#include <stdio.h>

{
    const char* path = "ch12_ex2.txt";

    // 1) Ghi đè bằng "w"
    FILE* fp = fopen(path, "w");
    if (!fp) {
        printf("open w failed\n");
    } else {
        fprintf(fp, "Lan 1\n");
        fclose(fp);

        // 2) Ghi thêm bằng "a"
        fp = fopen(path, "a");
        if (!fp) {
            printf("open a failed\n");
        } else {
            fprintf(fp, "Lan 2 (append)\n");
            fclose(fp);

            // 3) Đọc lại để xem kết quả
            fp = fopen(path, "r");
            if (!fp) {
                printf("open r failed\n");
            } else {
                char line[64];
                printf("Noi dung file:\n");
                while (fgets(line, sizeof(line), fp)) {
                    printf("%s", line);
                }
                fclose(fp);
            }
        }
    }
}
````

#### Ví dụ 3 — Đọc file theo dòng bằng `fgets` (loop đúng)
- Mục tiêu: pattern chuẩn: `while (fgets(...))`.


````python
#include <stdio.h>

{
    const char* path = "ch12_ex3.txt";

    // Tạo file mẫu
    FILE* fp = fopen(path, "w");
    if (!fp) {
        printf("open w failed\n");
    } else {
        fprintf(fp, "AAA\nBBB\nCCC\n");
        fclose(fp);

        // Đọc file
        fp = fopen(path, "r");
        if (!fp) {
            printf("open r failed\n");
        } else {
            char line[64];
            int lineNo = 0;
            while (fgets(line, sizeof(line), fp)) {
                lineNo++;
                printf("%d: %s", lineNo, line);
            }
            fclose(fp);
        }
    }
}
````

#### Ví dụ 4 — Binary I/O cơ bản (`fwrite`/`fread`)
- Mục tiêu: nhớ mode `wb/rb` và cách ghi/đọc dữ liệu nhị phân.


````python
#include <stdio.h>

{
    const char* path = "ch12_ex4.bin";

    int a[3] = {10, 20, 30};

    FILE* fp = fopen(path, "wb");
    if (!fp) {
        printf("open wb failed\n");
    } else {
        fwrite(a, sizeof(int), 3, fp);
        fclose(fp);

        int b[3] = {0};
        fp = fopen(path, "rb");
        if (!fp) {
            printf("open rb failed\n");
        } else {
            fread(b, sizeof(int), 3, fp);
            fclose(fp);

            printf("Read back: %d %d %d\n", b[0], b[1], b[2]);
        }
    }
}
````

## Chương 13. Toán tử bit

**Prerequisites**
- Chương 3 (toán tử).

**Learning outcomes**
- Phân biệt bitwise và logical: `&` vs `&&`, `|` vs `||`.
- Dùng dịch bit để tạo mask: `(1u << k)`.
- Thuộc công thức bitmask: set/clear/toggle/test.
- Nhận biết bẫy ưu tiên toán tử trong biểu thức bit.

### Giải thích

#### 1) Bitwise vs logical
- Bitwise `& | ^ ~` làm việc trên **từng bit**.
- Logical `&& || !` làm việc trên **đúng/sai** và có short-circuit.

Bẫy: dùng `&` trong `if` tưởng là `&&`.

#### 2) Dịch bit (shift) và mask
- `(1u << k)` tạo mask có bit k = 1.
- Nên dùng `unsigned` (`1u`) để tránh bẫy số âm.

#### 3) Công thức bitmask (cực hay thi)
Giả sử `mask = 1u << k`:
- Set bit k: `x |= mask`
- Clear bit k: `x &= ~mask`
- Toggle bit k: `x ^= mask`
- Test bit k: `(x & mask) != 0`

Bẫy precedence:
- Viết `(x & mask) == 0` nên có ngoặc.
- `x & mask == 0` bị hiểu sai thành `x & (mask == 0)`.

#### 4) XOR (optional)
- `a ^ a = 0`
- XOR swap có thể xuất hiện trong lý thuyết nhưng không khuyến khích viết thực tế.

### Ví dụ

#### Ví dụ 1 — Test / set / clear / toggle một bit
- Mục tiêu: thuộc 4 công thức bitmask.


````python
#include <stdio.h>

{
    unsigned int x = 0u;
    unsigned int k = 3u;
    unsigned int mask = 1u << k;

    printf("Ban dau: x = 0x%X\n", x);

    // Set bit k
    x |= mask;
    printf("Sau set: x = 0x%X\n", x);

    // Test bit k
    printf("Test bit %u: %s\n", k, ((x & mask) != 0u) ? "ON" : "OFF");

    // Toggle bit k
    x ^= mask;
    printf("Sau toggle: x = 0x%X\n", x);

    // Clear bit k
    x &= ~mask;
    printf("Sau clear: x = 0x%X\n", x);
}
````

#### Ví dụ 2 — Dịch bit để tạo mask
- Mục tiêu: nhớ `(1u << k)`.


````python
#include <stdio.h>

{
    unsigned int k = 5u;
    unsigned int mask = 1u << k;

    printf("k=%u => mask=0x%X\n", k, mask);

    // Bẫy: 1 << 31 (int) có thể overflow; dùng 1u để an toàn hơn.
}
````

#### Ví dụ 3 — Bẫy precedence: `x & mask == 0`
- Mục tiêu: thấy sự khác nhau giữa biểu thức sai và đúng.


````python
#include <stdio.h>

{
    unsigned int x = 1u;   // 0b...0001
    unsigned int mask = 2u; // 0b...0010

    // SAI: bị hiểu thành x & (mask == 0)
    unsigned int wrong = x & mask == 0u;

    // ĐÚNG: test bit bằng cách ngoặc rõ ràng
    unsigned int correct = ((x & mask) == 0u);

    printf("wrong   = %u\n", wrong);
    printf("correct = %u\n", correct);
}
````

#### Ví dụ 4 — XOR demo (optional)
- Mục tiêu: hiểu XOR: bit khác nhau → 1, giống nhau → 0.


````python
#include <stdio.h>

{
    unsigned int a = 5u; // 0101
    unsigned int b = 3u; // 0011

    printf("a^b = %u (0x%X)\n", a ^ b, a ^ b);
    printf("a^a = %u\n", a ^ a);
}
````

## Chương 17. Kế thừa và đa hình

**Prerequisites**
- Chương 15 (class).
- Chương 16 (new/delete).

**Learning outcomes**
- Hiểu quan hệ kế thừa (is-a) và cú pháp `class D : public B`.
- Hiểu `virtual` tạo đa hình động (runtime dispatch) khi gọi qua base pointer/reference.
- Phân biệt override vs overload (signature lệch là bẫy).
- Biết vì sao cần **virtual destructor** khi xóa qua base pointer.
- Nhận biết object slicing khi gán derived vào base by value.

### Giải thích

#### 1) Kế thừa (inheritance)
- `public` inheritance thể hiện quan hệ “D is-a B”.

#### 2) `virtual` và dynamic dispatch
- Nếu hàm base là `virtual` và derived override đúng signature, thì gọi qua `Base*` sẽ chạy phiên bản derived.

Bẫy:
- Quên `virtual` → gọi base.
- Signature lệch (khác `const`, khác tham số) → không override, chỉ overload.

#### 3) Virtual destructor
Nếu class có virtual method (đa hình), base destructor nên `virtual`.
- Nếu không, `delete basePtr` có thể không gọi destructor của derived (bẫy thi rất hay).

#### 4) Object slicing
- `Base b = Derived();` làm “cắt” phần derived.
- Đa hình chỉ hoạt động khi dùng pointer/reference.

### Ví dụ

#### Ví dụ 1 — Virtual dispatch qua `Base*`
- Mục tiêu: câu đề kinh điển: `Base* p = new D; p->f();`.


````python
#include <iostream>

#ifndef CH17_EX1_DEFINED
#define CH17_EX1_DEFINED
struct Base17 {
    virtual int f() { return 1; }
    virtual ~Base17() = default;
};

struct D17 : Base17 {
    int f() override { return 2; }
};
#endif

{
    Base17* p = new D17();
    std::cout << "p->f() = " << p->f() << "\n";
    delete p;
}
````

#### Ví dụ 2 — Override vs overload (signature lệch)
- Mục tiêu: bẫy thi: khác `const`/tham số => không override.
- Mẹo: dùng từ khóa `override` để compiler bắt lỗi.


````python
#include <iostream>

#ifndef CH17_EX2_DEFINED
#define CH17_EX2_DEFINED
struct Base17b {
    virtual int g() const { return 10; }
    virtual ~Base17b() = default;
};

struct D17b : Base17b {
    // KHONG override vì thiếu const (đây là overload/hiding)
    int g() { return 20; }

    // Nếu viết: int g() override { ... } thì compiler sẽ báo lỗi (tốt!).
};
#endif

{
    D17b d;

    Base17b* p = &d;
    std::cout << "Base* calls g() const -> " << p->g() << "\n";

    // Gọi trực tiếp bằng object D17b sẽ gọi bản không-const
    std::cout << "D object calls g() -> " << d.g() << "\n";
}
````

#### Ví dụ 3 — Virtual destructor: delete qua base pointer
- Mục tiêu: thấy thứ tự destructor: ~Derived rồi ~Base.


````python
#include <iostream>

#ifndef CH17_EX3_DEFINED
#define CH17_EX3_DEFINED
struct Base17c {
    virtual ~Base17c() { std::cout << "~Base\n"; }
};

struct D17c : Base17c {
    ~D17c() override { std::cout << "~Derived\n"; }
};
#endif

{
    Base17c* p = new D17c();
    delete p; // phải gọi ~Derived rồi ~Base
}
````

#### Ví dụ 4 — Object slicing (mất đa hình khi copy by value)
- Mục tiêu: nhớ: đa hình cần pointer/reference.


````python
#include <iostream>

#ifndef CH17_EX4_DEFINED
#define CH17_EX4_DEFINED
struct Base17d {
    virtual int f() { return 1; }
    virtual ~Base17d() = default;
};

struct D17d : Base17d {
    int f() override { return 2; }
};
#endif

{
    D17d d;

    Base17d b = d; // slicing: phần derived bị cắt
    std::cout << "b.f() = " << b.f() << " (khong con 2)\n";

    Base17d& ref = d;
    std::cout << "ref.f() = " << ref.f() << " (da hinh)\n";
}
````

## Chương 19. File I/O trong C++

**Prerequisites**

**Learning outcomes**
- Dùng `std::ifstream`/`std::ofstream`/`std::fstream` để đọc/ghi file.
- Check trạng thái stream (`if (!in)`, `is_open()`).
- Phân biệt đọc token (`>>`) và đọc dòng (`getline`).
- Dùng mode `std::ios::app` và `std::ios::binary`.
- Hiểu RAII: file tự đóng khi object ra khỏi scope.

### Giải thích

#### 1) Stream file trong C++
- `std::ifstream`: đọc.
- `std::ofstream`: ghi.
- `std::fstream`: đọc/ghi.

Luôn kiểm tra mở file:
- `if (!in)` hoặc `if (!in.is_open())`.

#### 2) `>>` vs `getline`
- `>>` đọc theo token (bỏ whitespace).
- `getline` đọc cả dòng (giữ khoảng trắng).

Bẫy newline giống Chương 14 khi trộn `>>` và `getline`.

#### 3) Mode
- `std::ios::app`: append.
- `std::ios::binary`: nhị phân.

#### 4) RAII auto-close
Object stream ra khỏi scope sẽ tự đóng file.

### Ví dụ

#### Ví dụ 1 — Copy file line-by-line (`getline` loop)
- Mục tiêu: pattern thực tế: while(getline(in,line)).


````python
#include <fstream>
#include <iostream>
#include <string>

{
    const char* inPath = "ch19_input.txt";
    const char* outPath = "ch19_output.txt";

    // Tạo file input mẫu
    {
        std::ofstream tmp(inPath);
        tmp << "Line 1\nLine 2\nLine 3\n";
    }

    std::ifstream in(inPath);
    std::ofstream out(outPath);

    if (!in || !out) {
        std::cerr << "Khong mo duoc file\n";
    } else {
        std::string line;
        while (std::getline(in, line)) {
            out << line << "\n";
        }

        std::cout << "Copied to " << outPath << "\n";
    }
}
````

#### Ví dụ 2 — Đọc số và tính tổng (`while(in >> x)`)
- Mục tiêu: loop đọc token chuẩn.


````python
#include <fstream>
#include <iostream>

{
    const char* path = "ch19_nums.txt";

    // Ghi file số
    {
        std::ofstream out(path);
        out << "1 2 3 4 5";
    }

    std::ifstream in(path);
    if (!in) {
        std::cerr << "Open failed\n";
    } else {
        int x;
        long long sum = 0;
        while (in >> x) {
            sum += x;
        }

        std::cout << "sum = " << sum << "\n";
    }
}
````

#### Ví dụ 3 — Bẫy newline khi trộn `>>` và `getline`
- Mục tiêu: nhớ dùng `std::ws` (hoặc ignore) trước getline.
- Ta mô phỏng bằng `istringstream` để chạy ổn định.


````python
#include <iostream>
#include <sstream>
#include <string>

{
    std::istringstream in("10\nHello file I/O\n");

    int n;
    std::string line;

    in >> n;
    std::getline(in, line);
    std::cout << "(wrong) line='" << line << "'\n";

    in.clear();
    in.seekg(0);
    in >> n;
    std::getline(in >> std::ws, line);
    std::cout << "(right) line='" << line << "'\n";
}
````

#### Ví dụ 4 — Binary write/read (ios::binary)
- Mục tiêu: nhớ mở mode binary khi đọc/ghi nhị phân.


````python
#include <fstream>
#include <iostream>

{
    const char* path = "ch19_bin.dat";

    int a[3] = {10, 20, 30};

    bool ok = true;

    // Write binary
    {
        std::ofstream out(path, std::ios::binary);
        if (!out) {
            std::cerr << "Open write failed\n";
            ok = false;
        } else {
            out.write(reinterpret_cast<const char*>(a), sizeof(a));
        }
    }

    int b[3] = {0};

    // Read binary
    if (ok) {
        std::ifstream in(path, std::ios::binary);
        if (!in) {
            std::cerr << "Open read failed\n";
            ok = false;
        } else {
            in.read(reinterpret_cast<char*>(b), sizeof(b));
        }
    }

    if (ok) {
        std::cout << b[0] << " " << b[1] << " " << b[2] << "\n";
    }
}
````

## Chương 20. Friend function/class

**Prerequisites**
- Chương 15 (class).

**Learning outcomes**
- Hiểu `friend` cho phép truy cập `private/protected` từ bên ngoài class.
- Biết `friend` không đối xứng và không kế thừa.
- Biết use-case hay thi: overload `operator<<` để in object.
- Biết khi nào nên dùng và khi nào tránh dùng `friend`.

### Giải thích

#### 1) `friend` là gì?
Khi bạn khai báo một hàm (hoặc class) là `friend` của một class:
- Hàm/class đó có quyền truy cập trực tiếp member `private/protected`.

Lưu ý quan trọng:
- `friend` **không** làm hàm trở thành member.
- `friend` **không đối xứng**: A friend B không có nghĩa B friend A.
- `friend` **không kế thừa**.

#### 2) Khi nào nên dùng?
Nên dùng khi:
- Cần một hàm ngoài class thao tác sâu với dữ liệu mà không muốn mở public quá nhiều.
- Thường gặp: `operator<<` để in object.

Tránh dùng khi:
- Bạn có thể giải quyết bằng public method/getter hợp lý.

### Ví dụ

#### Ví dụ 1 — Friend function truy cập private
- Mục tiêu: hiểu đúng “friend cho phép truy cập private”.


````python
#include <iostream>

#ifndef CH20_RECT_DEFINED
#define CH20_RECT_DEFINED
class Rect20 {
private:
    double w;
    double h;

public:
    Rect20(double w_, double h_) : w(w_), h(h_) {}

    // friend function được truy cập w,h
    friend double area(const Rect20& r);
};

double area(const Rect20& r) {
    return r.w * r.h;
}
#endif

{
    Rect20 r(3, 4);
    std::cout << "area = " << area(r) << "\n";
}
````

#### Ví dụ 2 — `operator<<` as friend (pattern rất hay thi)
- Mục tiêu: viết operator<< để `cout << obj` được.
- Bẫy: phải `return os;` để chain.


````python
#include <iostream>

#ifndef CH20_POINT_DEFINED
#define CH20_POINT_DEFINED
struct Point20 {
private:
    int x;
    int y;

public:
    Point20(int x_, int y_) : x(x_), y(y_) {}

    friend std::ostream& operator<<(std::ostream& os, const Point20& p);
};

std::ostream& operator<<(std::ostream& os, const Point20& p) {
    os << "(" << p.x << "," << p.y << ")";
    return os;
}
#endif

{
    Point20 p(1, 2);
    std::cout << "p = " << p << "\n";
}
````

#### Ví dụ 3 — Friend class (optional)
- Mục tiêu: thấy một class khác có thể truy cập private.
- Lưu ý: coupling mạnh, nên dùng hạn chế.


````python
#include <iostream>

#ifndef CH20_FRIENDCLASS_DEFINED
#define CH20_FRIENDCLASS_DEFINED
class B20; // forward declaration

class A20 {
private:
    int secret = 123;
    friend class B20;
};

class B20 {
public:
    void reveal(const A20& a) {
        std::cout << "secret = " << a.secret << "\n";
    }
};
#endif

{
    A20 a;
    B20 b;
    b.reveal(a);
}
````

## Mini-Mock theo chương

> Chạy cell Python ngay bên dưới để kích hoạt trắc nghiệm tương tác.


````python
import ipywidgets as w
from IPython.display import display, HTML
import html
import re
import json as _json
from pathlib import Path

RESULTS_PATH = Path.cwd() / '_mock_results.json'

STYLE = """
<style>
/* Force dark background around outputs in VS Code/Jupyter */
.output_container, .output, .output_area, .cell-output, .output_wrapper,
.jp-OutputArea-output, .jp-RenderedHTMLCommon, .vscode-notebook-cell-output,
.widget-area, .widget-subarea {
  background: #0b0b0b !important;
  color: #e6e6e6 !important;
}

.quiz-root { background: #0b0b0b; color: #e6e6e6; padding: 12px; border-radius: 8px; border: 1px solid #333; width: 100%; box-sizing: border-box; }
.quiz-topbar { margin: 4px 0 10px 0; }
.quiz-block { margin: 8px 0 18px 0; padding: 10px 12px; border: 1px solid #333; border-radius: 8px; background: #0b0b0b; color: #e6e6e6; }
.quiz-q { font-weight: 600; margin-bottom: 6px; color: #e6e6e6; }
.quiz-code { background: #111; color: #e6e6e6; padding: 10px; border-radius: 6px; border: 1px solid #333; }
.inline-code { background: #111; color: #ffd54f; padding: 1px 4px; border-radius: 4px; font-family: Consolas, Monaco, 'Courier New', monospace; }
.quiz-options { background: #0b0b0b; padding: 6px 8px; border-radius: 6px; border: 1px solid #222; }
.quiz-options .widget-label { color: #e6e6e6 !important; }
.quiz-options label { color: #e6e6e6 !important; }
.quiz-options input { accent-color: #22c55e; }
.quiz-close .widget-button { background: #dc2626 !important; color: #fff !important; border: 1px solid #7f1d1d !important; border-radius: 8px !important; }
.quiz-topbar .widget-togglebutton { border-radius: 8px !important; overflow: hidden; }
.quiz-topbar .widget-togglebutton > button { border-radius: 8px !important; }
.quiz-topbar .widget-button { border-radius: 8px !important; }
</style>
"""

def load_results():
    try:
        if RESULTS_PATH.exists():
            return _json.loads(RESULTS_PATH.read_text(encoding='utf-8'))
    except Exception:
        pass
    return {}

def save_results(data):
    try:
        RESULTS_PATH.write_text(_json.dumps(data, ensure_ascii=False, indent=2), encoding='utf-8')
    except Exception:
        pass

def render_inline_code(text: str) -> str:
    parts = text.split('`')
    out = []
    for i, part in enumerate(parts):
        if i % 2 == 0:
            out.append(html.escape(part))
        else:
            out.append(f"<code class='inline-code'>{html.escape(part)}</code>")
    return ''.join(out)

def clean_option_text(text: str) -> str:
    return re.sub(r'`([^`]*)`', r'\1', text)

def render_mock(title, questions):
    display(HTML(STYLE))
    results = load_results()
    key = title
    saved = results.get(key)

    header = w.HTML(f"<h4 style='margin: 0 0 8px 0; color:#e6e6e6'>{html.escape(title)}</h4>")

    mode = w.ToggleButtons(
        options=[('Chấm ngay', 'auto'), ('Làm hết rồi chấm', 'submit')],
        value='auto'
    )
    submit_btn = w.Button(description='Chấm và lưu', button_style='success', icon='check')
    submit_btn.layout.display = 'none'
    reset_btn = w.Button(description='Làm lại', button_style='warning', icon='refresh')
    reset_btn.layout.display = '' if saved else 'none'

    topbar = w.HBox([mode, submit_btn, reset_btn], layout=w.Layout(justify_content='space-between', align_items='center'))
    topbar.add_class('quiz-topbar')

    blocks = [header, topbar]
    items = []

    for q in questions:
        q_text = f"Câu {q['num']}. {q['q']}"
        q_html = w.HTML(f"<div class='quiz-q'>{render_inline_code(q_text)}</div>")
        parts = [q_html]

        if q.get('code'):
            style = q.get('code_style', 'block') or 'block'
            if style == 'inline':
                lines_inline = q['code'].splitlines() or [q['code']]
                html_lines = "".join([f"<code class='inline-code'>{html.escape(x)}</code><br>" for x in lines_inline if x != ""])
                parts.append(w.HTML(f"<div>{html_lines}</div>"))
            else:
                code_html = "<pre class='quiz-code'><code>" + html.escape(q['code']) + "</code></pre>"
                parts.append(w.HTML(code_html))

        opts = [(f"{k}. {clean_option_text(v)}", k) for k, v in q['options']]
        radio = w.RadioButtons(options=opts, value=None, layout=w.Layout(width='auto'))
        radio.add_class('quiz-options')
        fb = w.HTML("")

        def mark_answer(radio, fb, q):
            if radio.value is None:
                fb.value = "<span style='color:#ef4444'>Sai. Đáp án đúng: {}.</span>".format(q['answer'])
            elif radio.value == q['answer']:
                fb.value = "<span style='color:#22c55e'>Đúng.</span>"
            else:
                expl = q.get('explain','')
                expl = render_inline_code(expl) if expl else ''
                fb.value = f"<span style='color:#ef4444'>Sai. Đáp án đúng: {q['answer']}. {expl}</span>"

        def persist(graded=True):
            data = load_results()
            data[key] = {
                'mode': mode.value,
                'graded': graded,
                'answers': {str(q['num']): radio.value for radio, _, q in items}
            }
            save_results(data)

        def on_change(change, q=q, fb=fb, radio=radio):
            if change.get('name') != 'value':
                return
            if mode.value != 'auto':
                return
            if radio.value is None:
                fb.value = ""
            elif radio.value == q['answer']:
                fb.value = "<span style='color:#22c55e'>Đúng.</span>"
            else:
                expl = q.get('explain','')
                expl = render_inline_code(expl) if expl else ''
                fb.value = f"<span style='color:#ef4444'>Sai. Đáp án đúng: {q['answer']}. {expl}</span>"
            persist(graded=True)

        radio.observe(on_change)
        parts.extend([radio, fb])

        block = w.VBox(parts, layout=w.Layout(margin='6px 0 12px 0'))
        block.add_class('quiz-block')
        blocks.append(block)
        items.append((radio, fb, q))

    if saved:
        mode.value = saved.get('mode', 'auto')
        ans_map = saved.get('answers', {})
        for radio, fb, q in items:
            radio.value = ans_map.get(str(q['num']))
            mark_answer(radio, fb, q)
            radio.disabled = True
        mode.disabled = True
        submit_btn.disabled = True

    def on_submit(_):
        for radio, fb, q in items:
            mark_answer(radio, fb, q)
        persist(graded=True)
        reset_btn.layout.display = ''
        for radio, _, _ in items:
            radio.disabled = True
        mode.disabled = True
        submit_btn.disabled = True

    submit_btn.on_click(on_submit)

    def on_mode_change(change):
        if change.get('name') != 'value':
            return
        if mode.value == 'auto':
            submit_btn.layout.display = 'none'
        else:
            submit_btn.layout.display = ''
            for _, fb, _ in items:
                fb.value = ""

    mode.observe(on_mode_change)

    def on_reset(_):
        data = load_results()
        if key in data:
            data.pop(key)
            save_results(data)
        for radio, fb, _ in items:
            radio.value = None
            radio.disabled = False
            fb.value = ""
        mode.disabled = False
        mode.value = 'auto'
        submit_btn.disabled = False
        submit_btn.layout.display = 'none'
        reset_btn.layout.display = 'none'

    reset_btn.on_click(on_reset)

    close_btn = w.Button(description='Đóng mock', button_style='danger', icon='times')
    close_btn.layout = w.Layout(width='100%')
    close_btn.add_class('quiz-close')

    root = w.VBox(blocks + [close_btn], layout=w.Layout(width='100%', margin='0'))
    root.add_class('quiz-root')

    out = w.Output()
    def do_close(_):
        out.clear_output()
    close_btn.on_click(do_close)

    with out:
        display(root)
    display(out)
````
